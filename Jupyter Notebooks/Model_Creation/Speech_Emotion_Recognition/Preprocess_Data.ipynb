```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Emotions = pd.read_csv('/emotions.csv')
Emotions
```

```python
Emotions['Emotions'].unique()
```

```python
# Get the data
emotions_count = Emotions['Emotions'].value_counts()
emotions = emotions_count.index.to_list()
counts = emotions_count.to_list()

# Create the bar chart
plt.bar(emotions, counts)
plt.xlabel('Emotion')
plt.ylabel('Count')
plt.title('Bar Graph of Emotions')
plt.show()
```

```python
Emotions_new = Emotions[~Emotions['Emotions'].isin(['disgust', 'fear', 'surprise'])]
```

```python
Emotions_new
```

```python
Emotions_new['Emotions'].unique()
```

```python
# Get the data
emotions_count = Emotions_new['Emotions'].value_counts()
emotions = emotions_count.index.to_list()
counts = emotions_count.to_list()

# Create the bar chart
plt.bar(emotions, counts)
plt.xlabel('Emotion')
plt.ylabel('Count')
plt.title('Bar Graph of Emotions')
plt.show()
```

```python
# Create a mapping dictionary based on the 'Emotions' column for renaming
mapping = {'angry': 'Anger', 'neutral': 'Calmness', 'happy': 'Happiness', 'sad': 'Sadness'}

# Rename the columns using the mapping dictionary
Emotions_new["Emotions"] = Emotions_new['Emotions'].map(mapping)
```

```python
Emotions_new['Emotions'].unique()
```

```python
Emotions_new.to_csv('/Emotions_new.csv', index=False)
```
