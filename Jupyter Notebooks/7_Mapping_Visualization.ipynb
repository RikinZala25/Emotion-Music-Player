# Overall Similarity

```python
text_similarity = 6.11
acousic_similarity = 78.74
semantic_similarity = 53.3
```

```python
overall_similarity = (text_similarity+acousic_similarity+semantic_similarity)/3

overall_similarity
```

```python
import plotly.graph_objects as go
import numpy as np

# Create the data
data = [
    go.Scatterpolar(
        r=[text_similarity, acousic_similarity, semantic_similarity],
        theta=['Text', 'Acoustic', 'Semantic'],
        fill='toself',
        name='Similarity'
    )
]

# Create the layout
layout = go.Layout(
    title='Radar Map of Similarity',
    polar=dict(
        radialaxis=dict(
            range=[0, 100]
        )
    ),
    showlegend=True
)

# Create the figure
fig = go.Figure(data=data, layout=layout)

# Plot the figure
fig.show()
```

```python
audio_ser_output = ([{'label': 'Anger', 'confidence': 1.0},
  {'label': 'Calmness', 'confidence': 0},
  {'label': 'Happiness', 'confidence': 0},
  {'label': 'Sadness', 'confidence': 0}],
 [{'label': 'Anger', 'probability': 0.47536692},
  {'label': 'Calmness', 'probability': 0.17487772},
  {'label': 'Happiness', 'probability': 0.17487772},
  {'label': 'Sadness', 'probability': 0.17487772}])
```

```python
import plotly.graph_objects as go

labels = [x['label'] for x in audio_ser_output[0]]
values = [x['confidence'] for x in audio_ser_output[0]]

fig = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.35)])
fig.update_layout(title='Donut Chart of Audio Sentiment Confidence')
fig.show()
```
